# Rescaling Image and Video

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv.imread("Photos/dog2.jpg")
```

frame.shape[1] means image width, frame.shape[0] means image height

```python
def rescaleFrame(frame, scale=0.50):
    width = int(frame.shape[1] * scale)
    height = int(frame.shape[0] * scale)
    dimensions = (width, height)
    
    return cv.resize(frame, dimensions, interpolation=cv.INTER_AREA)
```

We are resizing it by 50%

### For Image

```python
resized_image = rescaleFrame(img)
#Show the image with matplotlib
RGB_img = cv.cvtColor(resized_image, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

### For Video

```python
capture = cv.VideoCapture("Videos/test_vid2.mp4")

while True:
    # Capture frame-by-frame
    isTrue, frame = capture.read()
    
    frame_resized = rescaleFrame(frame)
    
    # UNCOMMENT TO SEE THE VIDEO
    # cv.imshow('Resized frame', frame_resized)
    
    
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
