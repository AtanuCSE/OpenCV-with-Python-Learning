# Thresholding An Image

**Thresholding is basically converting an image pixels to 0 or 1**

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv.imread("Photos/dog2.jpg")
#Show the image with matplotlib
RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

## Simple Thresholding

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
threshold, thresh_img = cv.threshold(gray_img, 150, 255, cv.THRESH_BINARY)
#Show the image with matplotlib
RGB_img = cv.cvtColor(thresh_img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
# Inversed Threshold
# Changed the Threshold value to 170
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
threshold, thresh_inv_img = cv.threshold(gray_img, 170, 255, cv.THRESH_BINARY_INV)
#Show the image with matplotlib
RGB_img = cv.cvtColor(thresh_inv_img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

## Adaptive Thresholding
**Computer will decide suitable threshold value in this technique**

```python
thresh_img = cv.adaptiveThreshold(gray_img, 255, cv.ADAPTIVE_THRESH_MEAN_C, cv.THRESH_BINARY, 13, 5)
#Show the image with matplotlib
RGB_img = cv.cvtColor(thresh_img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
thresh_img = cv.adaptiveThreshold(gray_img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 13, 5)
#Show the image with matplotlib
RGB_img = cv.cvtColor(thresh_img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
