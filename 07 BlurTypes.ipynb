# Different Types of Blurring Techinques

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv.imread("Photos/dog2.jpg")
#Show the image with matplotlib
RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
# Averaging
average_blur_img = cv.blur(img, (7,7))
#Show the image with matplotlib
RGB_img = cv.cvtColor(average_blur_img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
# Gaussian Blur
gauss_blur_img = cv.GaussianBlur(img, (7,7), 0)
#Show the image with matplotlib
RGB_img = cv.cvtColor(gauss_blur_img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
# Median Blur
median_blur_image = cv.medianBlur(img, 7)
#Show the image with matplotlib
RGB_img = cv.cvtColor(median_blur_image, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
# Bilateral
bilateral_blur_image = cv.bilateralFilter(img, 10, 35, 25)
#Show the image with matplotlib
RGB_img = cv.cvtColor(bilateral_blur_image, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
