# Drawing Shape on Image

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
blank = np.zeros((500,500,3), dtype='uint8')
#Show the image with matplotlib
plt.imshow(blank)
plt.show()
```

```python
# Draw a Rectangle
withRec = cv.rectangle(blank, (100,50), (200, 300), (0,100,100), thickness=3)
#Show the image with matplotlib
plt.imshow(withRec)
plt.show()
```

**thickness=-1 to fill the shape with color**

```python
img = cv.imread("Photos/dog2.jpg")
# Draw A circle
circleOnImage = cv.circle(img, (640, 180), 150, (0,0,255), thickness=3)
#Show the image with matplotlib
circleOnImage = cv.cvtColor(circleOnImage, cv.COLOR_BGR2RGB)
plt.imshow(circleOnImage)
plt.show()
```

```python
img = cv.imread("Photos/dog2.jpg")
# Draw A circle
circleOnImage = cv.circle(img, (640, 180), 150, (0,0,255), thickness=3)
# Write text
circleOnImage = cv.putText(circleOnImage, 'I am a dog!!!', (800,180), cv.FONT_HERSHEY_COMPLEX, 1.5, (0,255,0), 3)
#Show the image with matplotlib
circleOnImage = cv.cvtColor(circleOnImage, cv.COLOR_BGR2RGB)
plt.imshow(circleOnImage)
plt.show()
```
