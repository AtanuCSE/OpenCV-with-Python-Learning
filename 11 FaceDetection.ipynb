# Face Detection using Built-in Libraries

**Haarcascade Classifier**

```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv.imread("Photos/person1.jpg")

#Show the image with matplotlib
RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(6,10))
plt.imshow(RGB_img)
plt.show()
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
#Show the image with matplotlib
plt.figure(figsize=(6,10))
plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
# Load haar cascade classifier

haar_cascade = cv.CascadeClassifier('Classifier/haar_face.xml')
```

```python
# Detect all faces

faces_rect = haar_cascade.detectMultiScale(gray_img, scaleFactor=1.1, minNeighbors=1)

print(f'Number of faces found = {len(faces_rect)}')
```

```python
for (x,y,w,h) in faces_rect:
    extracted_img = cv.rectangle(img, (x,y), (x+w,y+h), (0,255,0), thickness=2)
    
#Show the image with matplotlib
RGB_img = cv.cvtColor(extracted_img, cv.COLOR_BGR2RGB)
plt.figure(figsize=(6,10))
plt.imshow(RGB_img)
plt.show()
```
