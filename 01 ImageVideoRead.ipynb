# OpenCV Tutorial
## Basic Image Show

__Load OpenCV Library__

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

__Read an image__

```python
img = cv.imread("Photos/dog2.jpg")
```

cv.imshow has issues in jupyter

```python
#Show the image with matplotlib
RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
