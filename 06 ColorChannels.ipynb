# Split Image into Blue, Green, Red channels

**This may help to identify an object based on it's color. For example: field can be segmented out from the image with this manipulation**

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv.imread("Photos/dog2.jpg")
#Show the image with matplotlib
RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
b,g,r = cv.split(img)
```

### Dark pixels mean that less amount of that color is present in that area. For example: if we show the green split image then field would contain bright pixels. For blue, sky would contain more bright pixels.

```python
#Show the image with matplotlib
RGB_img = cv.cvtColor(b, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
#Show the image with matplotlib
RGB_img = cv.cvtColor(g, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
#Show the image with matplotlib
RGB_img = cv.cvtColor(r, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
