# Bitwise Operation and Use of Masking

## Bitwise Operation

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
blank = np.zeros((400,400), dtype='uint8')

rectangle = cv.rectangle(blank.copy(), (30,30), (370,370), 255, -1)
circle = cv.circle(blank.copy(), (200,200), 200, 255, -1)
```

```python
#Show the image with matplotlib
RGB_img = cv.cvtColor(rectangle, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
RGB_img = cv.cvtColor(circle, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
# bitwise AND --> intersecting regions
bitwise_and = cv.bitwise_and(rectangle, circle)
RGB_img = cv.cvtColor(bitwise_and, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
# bitwise XOR --> non-intersecting regions
bitwise_xor = cv.bitwise_xor(rectangle, circle)
RGB_img = cv.cvtColor(bitwise_xor, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

## Masking

```python
img = cv.imread("Photos/dog2.jpg")
RGB_img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
blank = np.zeros(img.shape[:2], dtype='uint8')
circle_mask = cv.circle(blank.copy(), (img.shape[1]//2 + 55,img.shape[0]//2), 200, 255, -1)
RGB_img = cv.cvtColor(circle_mask, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
masked_image = cv.bitwise_and(img,img, mask=circle_mask)
RGB_img = cv.cvtColor(masked_image, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
