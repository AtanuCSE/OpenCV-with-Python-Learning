# Edge Detection

- Canny
- Laplacian
- Sobel

```python
import cv2 as cv
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv.imread("Photos/dog5.jpg")
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
#Show the image with matplotlib
plt.imshow(gray_img, cmap='gray')
plt.show()
```

## Canny

```python
canny = cv.Canny(gray_img, 100, 175)
#Show the image with matplotlib
RGB_img = cv.cvtColor(canny, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

## Laplacian

```python
# Laplacian
lap = cv.Laplacian(gray_img, cv.CV_64F)
lap = np.uint8(np.absolute(lap))
#Show the image with matplotlib
RGB_img = cv.cvtColor(lap, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

## Sobel

```python
# Sobel 
sobelx = cv.Sobel(gray_img, cv.CV_64F, 1, 0)
sobely = cv.Sobel(gray_img, cv.CV_64F, 0, 1)
combined_sobel = cv.bitwise_or(sobelx, sobely)

sobelx = np.uint8(np.absolute(sobelx))
sobely = np.uint8(np.absolute(sobely))
combined_sobel = np.uint8(np.absolute(combined_sobel))


#Show the image with matplotlib
RGB_img = cv.cvtColor(sobelx, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()

RGB_img = cv.cvtColor(sobely, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()

RGB_img = cv.cvtColor(combined_sobel, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
