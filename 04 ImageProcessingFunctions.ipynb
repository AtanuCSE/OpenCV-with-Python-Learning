# Few Important Image Processing Functions

- RGB to Gray
- Blurred
- Finding edges in an image
- Sharpening the image / Dialating

```python
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv.imread("Photos/dog2.jpg")
# Converting to grayscale
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.imshow(gray_img, cmap='gray')
plt.show()
```

```python
img = cv.imread("Photos/dog2.jpg")
blur = cv.GaussianBlur(img, (7,7), cv.BORDER_DEFAULT)
#Show the image with matplotlib
RGB_img = cv.cvtColor(blur, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
img = cv.imread("Photos/dog2.jpg")
# Edge Cascade / Finding edges in the image
canny = cv.Canny(img, 200, 175)
#Show the image with matplotlib
RGB_img = cv.cvtColor(canny, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

**Blurring helps reducing noice in finding edges**

```python
# Edge Cascade / Finding edges in the image
canny_blurred = cv.Canny(blur, 125, 175)
#Show the image with matplotlib
RGB_img = cv.cvtColor(canny_blurred, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
# Dilating the image
dilated = cv.dilate(canny, (3,3), iterations=3)
#Show the image with matplotlib
RGB_img = cv.cvtColor(dilated, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```

```python
# Dilating the BLURRED image
dilated = cv.dilate(canny_blurred, (7,7), iterations=3)
#Show the image with matplotlib
RGB_img = cv.cvtColor(dilated, cv.COLOR_BGR2RGB)
plt.imshow(RGB_img)
plt.show()
```
